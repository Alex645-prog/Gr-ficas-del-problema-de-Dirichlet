### PROBLEMA 1

La solución de la ecuación: 

$$\dfrac{\partial^2 u   }{\partial x^2} + \dfrac{\partial^2 u   }{\partial y^2}=0,\,\,\, 0<x<1,~0<y<1$$

sujeta a las siguientes condiciones de frontera

$$u(0,y)=0, u(1,y)=0,\,\,\,x\in (0,1)\\ u(x,0)=100, u(x,1)=200,\,\,\,y\in (0,1)$$

Está dada en términos de la serie de Fourier por la expresión:

$$u(x,y)=\sum_{n=1}^{N} \left[E_n\sinh{(n \pi y)}+ F_n\cosh{(n\pi y)}     \right]\sin{(n\pi x)}     $$

donde:

$$F_n=\dfrac{200 \left( {{\left( -1\right) }^{n}}-1\right) }{\pi n}$$

y

$$E_n=\dfrac{\left( 200 {{\left( -1\right) }^{n}}-200\right)  \cosh{\left( \pi n\right) }-400 {{\left( -1\right) }^{n}}+400}{\pi  n \sinh{\left(\pi  n\right) }}$$

Mediante el siguiente código graficamos la solución a la EDP anterior

```octave
figure('position', [0, 0, 900, 600]);
nx = 100;
ny = 100;
x = linspace(0, 1, nx);
y = linspace(0, 1, ny);
[X, Y] = meshgrid(x, y);

N = 10; # El número de términos a utilizar en la sumatoria

u = zeros(nx, ny);
for n = 1:N
    F_n = -(200*((-1)^n-1))/(pi*n);
    E_n = ((200*(-1)^n-200)*cosh(pi*n)-400*(-1)^n+400)/(pi*n*sinh(pi*n));
    u += (E_n*sinh(n*pi*Y) + F_n*cosh(n*pi*Y))*sin(n*pi*X);
end

u(1,:) = 0;
u(nx,:) = 0;
u(:,1) = 100;
u(:,ny) = 200;

surf(X, Y, u);
colormap(jet);
shading interp;
xlabel('x');
ylabel('y');
title("u(x,t)")

```
