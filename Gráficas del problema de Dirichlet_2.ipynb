La solución de la ecuación: 

$$\dfrac{\partial^2 u   }{\partial x^2} + \dfrac{\partial^2 u   }{\partial y^2}=0,\,\,\, 0<x<1,~0<y<1$$

sujeta a las siguientes condiciones de frontera

$$u(0,y)=10y,\,\,\,\left. \dfrac{\partial u}{\partial x} \right|_{x=1}=-1\\ u(x,0)=0,\,\,\,u(x,1)=0$$

Está dada en términos de la serie de Fourier por la expresión:

$$u(x,y)=\sum_{n=1}^{N} \left[E_n\sinh{(n \pi x)}+ F_n\cosh{(n\pi x)}     \right]\sin{(n\pi y)}     $$

donde

$$F_n=-\dfrac{20 {{\left( -1\right) }^{n}}}{\pi  n}$$

y

$$E_n=\dfrac{20 {{\left( -1\right) }^{n}} \sinh{\left( \pi  n\right) }+2 \pi  n {{\left( -1\right) }^{n}}-2 \pi  n}{\pi  n \cosh{\left( \pi  n\right) }}$$

Mediante el siguiente código graficamos la solución a la EDP anterior

```octave
figure('position', [0, 0, 900, 600]);


nx = 100;
ny = 100;
x = linspace(0, 1, nx);
y = linspace(0, 1, ny);
[X, Y] = meshgrid(x, y);

N = 5; # El número de términos a utilizar en la sumatoria

u = zeros(nx, ny);
for n = 1:N
    F_n = -(20*(-1)^n)/(pi*n);
    E_n = (20*(-1)^n*sinh(pi*n)+2*pi*n*(-1)^n-2*pi*n)/(pi*n*cosh(pi*n));
    u += (E_n*sinh(n*pi*X) + F_n*cosh(n*pi*X))*sin(n*pi*Y);
end
% Agregar condiciones de frontera
u(:,1) = 0;
u(:,end) = 10*y;
u(1,:) = 0;
u(end,:) = -1;
surf(X, Y, u);
colormap(jet);
shading interp;
xlabel('x');
ylabel('y');
title("u(x,t)")
```

```octave
figure('position', [0, 0, 900, 600]);

nx = 100;
ny = 100;
x = linspace(0, 1, nx);
y = linspace(0, 1, ny);
[X, Y] = meshgrid(x, y);

N = 10; # El número de términos a utilizar en la sumatoria

u = zeros(nx, ny);
for n = 1:N
    F_n = -(20*(-1)^n)/(pi*n);
    E_n = 2*sech(n*pi)*((-1)^n-1)-F_n*tanh(n*pi);
    u += (E_n*sinh(n*pi*X) + F_n*cosh(n*pi*X))*sin(n*pi*Y);
end

imagesc(x, y, u);
colormap(jet);
colorbar;
xlabel('x');
ylabel('y');
title("u(x,t)")
```
